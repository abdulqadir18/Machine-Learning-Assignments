```python
# https://www.kaggle.com/datasets/datascientistanna/customers-dataset
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler, normalize
from sklearn.metrics import silhouette_score
import scipy.cluster.hierarchy as shc
```

```python
# getting csv file
df = pd.read_csv('Customers.csv')
df
```

```python
# checking NA values
df.isna().sum()
```

```python
# dropping NA values
df.dropna(inplace=True)
```

```python
# Scatter Plot
plt.grid(True)
plt.scatter(df['CustomerID'],df['Age'])
plt.show()
```

```python
# histogram
plt.hist(df['Annual Income ($)'], bins=[0, 50000, 100000, 150000, 200000, 250000])
```

```python
# count plot
sns.countplot(x = 'Profession', data = df)
```

```python
# box plot
sns.boxplot(x = 'Work Experience', data = df)
```

```python
# Label Encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Profession']= label_encoder.fit_transform(df['Profession'])
```

```python
# get object columns
col = df.select_dtypes(include = 'object').columns
# One Hot Encoding
df = pd.get_dummies(df, columns = col)
df = df.drop('CustomerID', axis=1)
df.isna().sum()
```

```python
x = df
```

```python
# Standard Scaler
# from sklearn.preprocessing import StandardScaler 
# scalar = StandardScaler()
# scalar.fit(x)
# x = scalar.transform(x)
```

```python
# Ward
plt.figure(figsize =(8, 8))
plt.title('Visualising the data')
Dendrogram = shc.dendrogram((shc.linkage(x, method ='ward')))
```

```python
ac1 = AgglomerativeClustering(n_clusters = 2, compute_full_tree='auto', metric='euclidean', linkage='ward')
silhouette_score(x, ac1.fit_predict(x))
```

```python
# Complete
plt.figure(figsize =(8, 8))
plt.title('Visualising the data')
Dendrogram = shc.dendrogram((shc.linkage(x, method ='complete')))
```

```python
ac2 = AgglomerativeClustering(n_clusters = 2, compute_full_tree='auto', metric='euclidean', linkage='complete')
silhouette_score(x, ac2.fit_predict(x))
```

```python
# Average
plt.figure(figsize =(8, 8))
plt.title('Visualising the data')
Dendrogram = shc.dendrogram((shc.linkage(x, method ='average')))
```

```python
ac3 = AgglomerativeClustering(n_clusters = 2, compute_full_tree='auto', metric='euclidean', linkage='average')
silhouette_score(x, ac3.fit_predict(x))
```

```python
# Single
plt.figure(figsize =(8, 8))
plt.title('Visualising the data')
Dendrogram = shc.dendrogram((shc.linkage(x, method ='single')))
```

```python
ac4 = AgglomerativeClustering(n_clusters = 2, compute_full_tree='auto', metric='euclidean', linkage='single')
silhouette_score(x, ac4.fit_predict(x))
```
