```python
# https://www.kaggle.com/code/karanchinchpure/bank-marketing-who-will-subscribe-for-deposit
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading csv file
df = pd.read_csv('bank.csv')
```

```python
df.head()
```

```python
# handling na values
df.isna().count

# Replace na values if there are any
df.fillna(df.mean(), inplace=True)
df = df.fillna(df.mode().iloc[0])
```

```python
# scatter plot
x1 = df['balance']
x2 = df['deposit']
plt.scatter(x1, x2)
```

```python
# histogram
plt.hist(x1, bins=[0, 5, 10, 15, 20, 25])
```

```python
# count plot
sns.countplot(x = 'deposit', data = df)
```

```python
# box plot
sns.boxplot(x = 'balance', data = df)
```

```python
# get object columns
col = df.select_dtypes(include = 'object').columns
# # One Hot Encoding
# x = pd.get_dummies(x, columns = col)
# x.head()

# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in col:
    df[feat] = le.fit_transform(df[feat].astype(str))
```

```python
# correlation matrix
sample1 =df.iloc[: , 1 : 8].copy()
sample1['deposit'] = df['deposit']
sample2 =df.iloc[: , 9 : 16].copy()
sample2['deposit'] = df['deposit']
sns.heatmap(sample1.corr(), annot = True, cmap = 'RdBu')
```

```python
sns.heatmap(sample2.corr(), annot = True, cmap = 'RdBu')
```

```python
# dropping features
df = df.drop(['default', 'month', 'day' ], axis=1)
```

```python
# Splitting features and target
y= df['deposit']
x= df.drop('deposit', axis=1)
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler 
scalar = StandardScaler()
scalar.fit(x)
x = scalar.transform(x)
```

```python
# Splitting in training testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
# Applying SVM rbf
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Predicting on test
y_pred = classifier.predict(X_test) 
y_pred
```

```python
# Accuracy
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```

```python
# Applying SVM sigmoid
from sklearn.svm import SVC
classifier = SVC(kernel = 'sigmoid', random_state = 0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test) 
metrics.accuracy_score(y_test, y_pred)
```

```python
# Applying SVM poly
from sklearn.svm import SVC
classifier = SVC(kernel = 'poly', random_state = 0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test) 
metrics.accuracy_score(y_test, y_pred)
```

```python
# Applying SVM linear
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test) 
metrics.accuracy_score(y_test, y_pred)
```

```python
# Confusion Metrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, xticklabels=['Yes','No'], yticklabels=['Yes','No'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.show()
```
