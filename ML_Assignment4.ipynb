```python
# import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv
df = pd.read_csv('Train_Dataset.csv')
test_df = pd.read_csv('Test_Dataset.csv')
```

```python
# deal with na values if any
df.isna()
```

```python
# replace na values
df.fillna(df.mean(), inplace=True)
test_df.fillna(test_df.mean(), inplace=True)
df = df.fillna(df.mode().iloc[0])
test_df = test_df.fillna(test_df.mode().iloc[0])

df.isna()
```

```python
# Label Encoding
# 1. find columns with object
# 2. apply label encoder in loop
col = df.select_dtypes(include = 'object').columns
col 
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in col:
    df[feat] = le.fit_transform(df[feat].astype(str))
```

```python
test_col=test_df.select_dtypes(include='object').columns

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in col:
    test_df[feat] = le.fit_transform(test_df[feat].astype(str))
```

```python
# Splitting in  training and testing data
x_train, y_train = df.drop('Default', axis=1), df['Default']
# x_test, y_test = test_df.drop('Default', axis=1), test_df['Default']
```

```python
# importing libraries for neural network
from keras import models
from keras import layers
```

```python
# creating neural network
network = models.Sequential()
# initial input size
network.add(layers.Dense(512, activation='relu'))
# subsequent hidden layer
network.add(layers.Dense(256, activation='relu'))
# subsequent hidden layer
network.add(layers.Dense(256, activation='relu'))
# final output (in our case 0-9 image)
network.add(layers.Dense(1, activation='sigmoid'))
```

```python
# gradient, loss function, metrics
network.compile(optimizer='rmsprop', # same as ngradient
loss = 'binary_crossentropy',   # loss function is cross entropy
metrics=['accuracy'])                # evaluation matrix is accuracy
```

```python
# Our vectorized labels
y_train = np.asarray(y_train).astype('float32').reshape((-1,1))
```

```python
# epochs is no of times it should run
network.fit(x_train, y_train, epochs=10, batch_size=64, validation_split=0.2)
```
