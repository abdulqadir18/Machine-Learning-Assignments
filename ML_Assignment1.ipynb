```python
# One Hot Encoding
col = df.select_dtypes(include = 'object').columns
col
```

```python
df = pd.get_dummies(df, columns = col)
# df = pd.get_dummies(df, columns = ['MSZoning'])
df.head()
```

```python
test_col=test_df.select_dtypes(include='object').columns
test_col
test_df = pd.get_dummies(test_df, columns = test_col)
```

```python
# Splitting features and target
y= df['SalePrice']
x= df.drop('SalePrice', axis=1)
# inner join to handle new columns in test
final_train, final_test = x.align(test_df, join='inner', axis=1) 
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin = lin.fit(final_train,y)
y_pred = lin.predict(final_train)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y,y_pred))
```
